```python
from mlxtend.frequent_patterns import apriori
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from sklearn.preprocessing import LabelEncoder

```

```python
#using association ruls algorithm to realworld data
```

```python
#explor the data
df =pd.read_csv(r"C:\Users\DellPc\Desktop\qui.csv")
df
```

```python
#we have to drop nan value from the data
df.dropna(axis=0,how="any",inplace=True)
df
```

```python
#transform the order_id column to numaric
le=LabelEncoder()
df["order_id"]=le.fit_transform(df["order_id"])
df
```

```python
#we have to tranform the shape of data to Appropriate shape 
# here we used the The method suggested by AI 
df_one= df.groupby(["order_id"])["product_title"].apply(list)
df_one
```

```python
df_one= df_one.apply(pd.Series).stack().reset_index(level=1,drop=True).to_frame("product_title")
df_one
```

```python
df_one = df_one.reset_index().drop_duplicates().pivot(index="order_id",columns="product_title",values="product_title").fillna(0).astype(bool)
```

```python
df_one
```

```python
#here we apply the algorithm to data 
freq_item= apriori(df_one,min_support=0.01,use_colnames=True) 
rules=association_rules(freq_item,metric="lift",min_threshold=0.5)
```

```python
rules
```

```python
#creating function to determine the items that appropriate to cross sell model
def recommend_by_item(product_title,rules,df_one):
    trans=df_one[df_one.loc[:,product_title]==1].index
    recommendations=[]
    for tran in trans:
        
        for antecedent,consequent,support,confidence,lift in zip(rules["antecedents"], rules["consequents"], rules["support"], rules["confidence"], rules["lift"]):
            if set([product_title]).issubset(antecedent):
                recommendations.append(consequent.difference(trans))
    return recommendations
    
```

```python
r=recommend_by_item("Sheyla cotton Instant Hijab",rules,df_one)
```

```python
recommendations_str = []
for item_set in recommendations:
    item_str = ', '.join(item_set)
    recommendations_str.append(item_str)
    
```

```python
r_data=pd.DataFrame(r)
print(r_data.drop_duplicates())
```

```python
dfw= pd.read_csv(r"C:\Users\DellPc\Downloads\aug_train\aug_train.csv")
```

```python
dfw= dfw.drop(columns=['id','Policy_Sales_Channel','Vintage',"Region_Code"])
columns=["Gender","Vehicle_Damage","Driving_License","Previously_Insured"]
for i in columns:
    dfw=pd.get_dummies(dfw,columns=[i],prefix=i)
dfw["Age"]=pd.cut(dfw["Age"],bins=[0, 29, 35, 50, 100])
dfw["Age"]=dfw["Age"].cat.codes
dfw['Annual_Premium'] = pd.cut(dfw['Annual_Premium'], bins=[0, 30000, 35000,40000, 45000, 50000, np.inf])
dfw['Annual_Premium']= dfw['Annual_Premium'].cat.codes
dfw['Vehicle_Age'] =dfw['Vehicle_Age'].map({'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2})
```

```python
from sklearn.model_selection import train_test_split 
Features= ['Age','Vehicle_Age','Annual_Premium',"Gender_Female","Gender_Male","Vehicle_Damage_No","Vehicle_Damage_Yes",
"Driving_License_0","Driving_License_1" ,"Previously_Insured_0", "Previously_Insured_1"]
X_train, X_test, Y_train, Y_test = train_test_split(dfw[Features],dfw['Response'],
                                   test_size = 0.3, random_state = 101) 
X_train.shape,X_test.shape
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus=RandomUnderSampler(sampling_strategy=0.5,random_state=42)
X_train,Y_train =rus.fit_resample(dfw[Features],dfw["Response"])
```

```python
from sklearn.metrics import accuracy_score, f1_score,auc

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import RandomForestClassifier
```

```python
def performance_met(model,X_train,Y_train,X_test,Y_test):
    acc_train=accuracy_score(Y_train, model.predict(X_train))
    f1_train=f1_score(Y_train, model.predict(X_train))
    acc_test=accuracy_score(Y_test, model.predict(X_test))
    f1_test=f1_score(Y_test, model.predict(X_test))
    print("train score: accuracy:{} f1:{}".format(acc_train,f1_train))
    print("test score: accuracy:{} f1:{}".format(acc_test,f1_test))
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train) 
performance_met(model,X_train,Y_train,X_test,Y_test)
```

```python
model_DT=DecisionTreeClassifier(random_state=1) 
model_DT.fit(X_train,Y_train) 
performance_met(model_DT,X_train,Y_train,X_test,Y_test)
```

```python
Forest= RandomForestClassifier(random_state=1) 
Forest.fit(X_train,Y_train) 
performance_met(Forest,X_train,Y_train,X_test,Y_test)
```
